```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV data
data = pd.read_csv('prices_round_1_day_-2.csv', delimiter=';')

# Separate data by product
products = data['product'].unique()

# Create a subplot for each product
fig, axs = plt.subplots(len(products), 1, figsize=(10, 5*len(products)))

for i, product in enumerate(products):
    product_data = data[data['product'] == product]
    
    # Plot the data in its own subplot
    axs[i].plot(product_data['timestamp'], product_data['mid_price'])
    axs[i].set_title(product)
    axs[i].set_xlabel('Timestamp')
    axs[i].set_ylabel('Mid Price')

plt.tight_layout()
plt.show()
```

```python
# Read the CSV data from three different days
data_day_0 = pd.read_csv('prices_round_1_day_0.csv', delimiter=';')
data_day_n1 = pd.read_csv('prices_round_1_day_-1.csv', delimiter=';')
data_day_n2 = pd.read_csv('prices_round_1_day_-2.csv', delimiter=';')

# Add a constant to the timestamps of day -1 and day 0
data_day_n1['timestamp'] += max(data_day_n2['timestamp']) + 1
data_day_0['timestamp'] += max(data_day_n1['timestamp']) + 1

# Concatenate the data
data = pd.concat([data_day_n2, data_day_n1, data_day_0])

# Separate data by product
products = data['product'].unique()

# Create a subplot for each product
fig, axs = plt.subplots(len(products), 1, figsize=(10, 5*len(products)))

for i, product in enumerate(products):
    product_data = data[data['product'] == product]
    
    # Plot the data in its own subplot
    axs[i].plot(product_data['timestamp'], product_data['mid_price'])
    axs[i].set_title(product)
    axs[i].set_xlabel('Timestamp')
    axs[i].set_ylabel('Mid Price')

plt.tight_layout()
plt.show()
```
